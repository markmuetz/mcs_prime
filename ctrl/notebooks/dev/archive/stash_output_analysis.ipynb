```python
import pandas as pd
```

```python
df1 = pd.read_csv('/home/markmuetz/Datasets/MCS_PRIME/analysis/u-cp288_stash_output.csv', sep=',')
df2 = pd.read_csv('/home/markmuetz/Datasets/MCS_PRIME/analysis/u-cp308_stash_output.csv', sep=',')
```

```python
# Need to make sure there are no differences between the two dfs so I can compare them.
df1['file'] = df1.file.str.replace('cp288a', 'sim')
df2['file'] = df2.file.str.replace('cp308a', 'sim')
```

```python
# Get string for each row as a csv row.
rows1 = df1.to_csv(header=None, index=False).strip('\n').split('\n')
rows2 = df2.to_csv(header=None, index=False).strip('\n').split('\n')
```

```python
print(len(rows1))
print(len(rows1) == len(set(rows1)))
print(len(rows2))
print(len(rows2) == len(set(rows2)))
```

```python
len(set(rows1) & set(rows2))
```

i.e. all rows in rows1 are in both.

```python
set(rows2) - set(rows1)
```

All these look like they are related to convection -- good! This means that the simulation with CoMorph (u-cp288) is not outputting them because they are not present for CoMorph.

```python
# Way of getting at same info as above using just pandas.
# https://stackoverflow.com/a/47107164/54557
df_all = df2.merge(df1.drop_duplicates(), on=[v for v in df1.columns.values],
                   how='left', indicator=True)
df2_only = df_all[df_all._merge == 'left_only']
```

```python
df2_only
```

```python
(df2_only.stash_code.str.split(':', expand=True).values[:, 0] == '5').all()
```

And indeed, all values come from the convection scheme (section 5).
