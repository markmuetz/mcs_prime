```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(20, 60, 41)
y = np.linspace(-20, 20, 41)
X, Y = np.meshgrid(x, y)
```

```python
def exp_hill(X, Y, xloc, yloc, xscale=5, yscale=5):
    return np.exp(-(((X - xloc)**2) / xscale + ((Y - yloc)**2) / yscale))

Z = exp_hill(X, Y, 30, 10) - 3 * exp_hill(X, Y, 45, 5, xscale=10)
```

```python
plt.imshow(Z)
```

```python
plt.contourf(x, y, Z)
plt.colorbar()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})
im = ax.contourf(x, y, Z, vmin=-3, vmax=3)
ax.coastlines()
plt.colorbar(im)
```

```python
levels=np.linspace(-3, 3, 10)
levels=[-3, -2, -1, -0.5, 0.5, 1, 2, 3]
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})
im = ax.contourf(x, y, Z, levels=levels, cmap='bwr')
ax.coastlines()
plt.colorbar(im)
```
