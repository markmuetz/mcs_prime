Chen, Liu and Mapes (2017) define 2 useful RH values, both of which have predictive value. RHlow: avg over 1000-850 hPa layer. RHmid: avg over 700-400 hPa. Calc these below:

```python
# 114: 856 hPa - https://confluence.ecmwf.int/display/UDOC/L137+model+level+definitions
RHlow = e5calc.calc_RH(p[114:], ds.t.values[0][114:], ds.q.values[0][114:]).mean(axis=0)
```

```python
# 90: 407 hPa - https://confluence.ecmwf.int/display/UDOC/L137+model+level+definitions
# 105: 703 hPa - https://confluence.ecmwf.int/display/UDOC/L137+model+level+definitions
RHmid = e5calc.calc_RH(p[90:106], ds.t.values[0][90:106], ds.q.values[0][90:106]).mean(axis=0)
```

Chen et al. (2023) show that theta_e has the largest impact just before MCS init in layer 850-750 hPa (their Fig. 1).
Use this to define theta_e layer: theta_e_mid

```python
# 107: 742 hPa
# 114: 857 hPa
theta_e_mid = e5calc.calc_theta_e(p[107:115], ds.t.values[0][107:115], ds.q.values[0][107:115]).mean(axis=0)
theta_e_mid
```
