```python
def display_grid(nrow, ncol, paths, width=10):
    imgs = [plt.imread(p) for p in paths]
    h = imgs[0].shape[0]
    w = imgs[0].shape[1]
    scale = w / width
    fig, axes = plt.subplots(2, 2, figsize=(w / scale, h / scale))
    for ax, img in zip(axes.flatten(), imgs):
        ax.imshow(img)
        ax.axis('off')
    plt.subplots_adjust(0, 0, 1, 1, 0, 0)
    plt.show()
```
