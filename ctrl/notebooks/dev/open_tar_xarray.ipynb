```python
import tarfile

import xarray as xr
```

```python
tar = tarfile.TarFile('/gws/nopw/j04/mcsprime/mmuetz/data/MCS_Global/mcstracking/20000601.0000_20010101.0000.tar')
```

```python
# tar.getmembers()
```

```python
with tar.extractfile('20000601.0000_20010101.0000/mcstrack_20001211_2030.nc') as flike:
    ds = xr.load_dataset(flike, engine='h5netcdf')
ds
```

```python
datasets = []
for h in range(24):
    print(h)
    # N.B. a `with tar.extractfile(...) as tf:` will not work here, as
    # `xr.concat` needs open files.
    # Loading the data uses too much memory.
    datasets.append(xr.open_dataset(
        tar.extractfile(f'20000601.0000_20010101.0000/mcstrack_20001211_{h:02d}30.nc'),
        engine='h5netcdf',
    ))
ds = xr.concat(datasets, dim='time')
```
