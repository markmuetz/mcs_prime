* Move to templates.

```python
import datetime as dt

import xarray as xr

from mcs_prime import PATHS, McsTracks, PixelData
```

```python
year = 2019
month = 6
day = 21
e5datadir = PATHS['era5dir'] / f'data/oper/an_ml/{year}/{month:02d}/{day:02d}'
```

```python
h = 6

e5time = dt.datetime(year, month, day, h, 0)
paths = [e5datadir / (f'ecmwf-era5_oper_an_ml_{t.year}{t.month:02d}{t.day:02d}'
                      f'{t.hour:02d}00.{var}.nc')
         for var in ['u', 'v']
         for t in [e5time, e5time + dt.timedelta(hours=1)]]
```

```python
e5uv = (xr.open_mfdataset(paths).sel(latitude=slice(60, -60))
        .mean(dim='time').load())
```

```python
lons = [25.321, 24.987, 21.1]
lats = [12.123, 21.981, 11.2]
lon = xr.DataArray(lons, dims='track_point')
lat = xr.DataArray(lats, dims='track_point')

e5uv.interp(longitude=lon, latitude=lat).u
```
