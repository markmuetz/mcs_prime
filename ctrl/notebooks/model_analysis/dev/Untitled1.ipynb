```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage as ndimage
```

```python
a = (np.random.random((100, 100)) > 0.9).astype(float)
```

```python
plt.imshow(a, interpolation='nearest')
```

```python
plt.imshow(ndimage.gaussian_filter(a, (0, 3)), interpolation='nearest')
```

```python
(a == ndimage.gaussian_filter(a, 0)).all()
```
