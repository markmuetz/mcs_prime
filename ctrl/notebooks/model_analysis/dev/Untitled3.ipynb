```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
```

```python
a = np.zeros((101, 101))
x = np.linspace(0, 100, 101)
y = np.linspace(0, 100, 101)
X, Y = np.meshgrid(x, y)
```

```python
sigma = 20
C = 1 / (sigma * (2 * np.pi)**0.5)
z = C * np.exp(-0.5 * ((X - 50)**2 + (Y - 50)**2) / sigma**2)
a[50, 50] = z.sum()
gfa = gaussian_filter(a, sigma=sigma)



print(a.sum())
print(gfa.sum())
print(z.sum())
```

```python
fig, (ax0, ax1, ax2, ax3) = plt.subplots(1, 4)
fig.set_size_inches((12, 3))
ax0.imshow(a)
ax1.imshow(gfa, vmin=0, vmax=C)
ax2.imshow(z, vmin=0, vmax=C)
ax3.imshow(z - gfa, vmin=-C, vmax=C, cmap='bwr')
```
