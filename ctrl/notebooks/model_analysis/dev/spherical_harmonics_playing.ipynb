```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special
```

```python
N_UM = 96
x = np.linspace(0, 360, N_UM * 2 + 1)
y = np.linspace(0, 180, int(N_UM * 1.5) + 1)
X, Y = np.meshgrid(x, y)
```

```python
theta = X * np.pi / 180
phi = Y * np.pi / 180
```

```python
plt.imshow(scipy.special.sph_harm(2, 3, theta, phi).real, origin='lower', cmap='bwr')
```

```python
N = 6
fig, axes = plt.subplots(N, N)
fig.set_size_inches((10, 10))
for m in range(N):
    for n in range(N):
        ax = axes[m, n]
        ax.imshow(scipy.special.sph_harm(m, n, theta, phi).real, origin='lower', cmap='bwr')
```
